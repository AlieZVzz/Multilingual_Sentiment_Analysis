```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
plt.rcParams['font.sans-serif']=['Simhei']
```

### 中文Acc曲线

```python
attention_acc = np.load('save/BiLSTM/acc.npy')
textcnn_acc = np.load('save/CNN/acc.npy')
plt.plot(attention_acc)
plt.plot(textcnn_acc)
plt.xlabel('训练轮次')
plt.xticks(range(0,20,2))
plt.ylabel('验证集准确率')
plt.grid()
plt.legend(['BiLSTM','TextCNN'])
plt.savefig('Chinese Accuracy.png', dpi=300, facecolor='white')
```

### 中文Loss曲线

```python
attention_loss = np.load('save/BiLSTM/loss.npy')
textcnn_loss = np.load('save/CNN/loss.npy')
plt.plot(attention_loss)
plt.plot(textcnn_loss)
plt.xlabel('训练轮次')
plt.xticks(range(0,20,2))
plt.ylabel('验证集损失')
plt.legend(['BiLSTM','TextCNN'])
plt.grid()
plt.savefig('Chinese Loss.png', dpi=300,  facecolor='white')
```
